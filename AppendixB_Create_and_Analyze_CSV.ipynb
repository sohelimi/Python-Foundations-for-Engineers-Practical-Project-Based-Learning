# Appendix B: Creating and Analyzing a CSV File with Dummy Data

This appendix demonstrates how to create a CSV file with dummy data, save it, read it back, and perform statistical analysis with visualization.

## Step 1: Create and Save a CSV File

We create a dataset with five columns: `Date`, `KPI1`, `KPI2`, `KPI3`, `KPI4`, and `KPI5`. The data includes a range of random values representing Key Performance Indicators (KPIs).

```python
import pandas as pd
import numpy as np

# Generate dummy data
np.random.seed(42)
date_range = pd.date_range(start="2024-01-01", periods=30)
data = {
    "Date": date_range,
    "KPI1": np.random.randint(50, 100, size=30),
    "KPI2": np.random.randint(20, 80, size=30),
    "KPI3": np.random.randint(10, 60, size=30),
    "KPI4": np.random.uniform(1.5, 3.5, size=30),
    "KPI5": np.random.uniform(0.5, 1.5, size=30),
}

# Save the data to a CSV file
df = pd.DataFrame(data)
df.to_csv("dummy_kpi_data.csv", index=False)
print("CSV file created and saved!")
```

## Step 2: Read and Analyze the CSV File

We read the data from the CSV file and perform statistical analysis to compute key metrics such as mean, median, and standard deviation.

```python
# Read the CSV file
df = pd.read_csv("dummy_kpi_data.csv")

# Display the first few rows
print(df.head())

# Statistical summary
print("Statistical Summary:")
print(df.describe())
```

## Step 3: Visualize the Data

We use visualizations to understand trends and distributions in the data.

```python
import matplotlib.pyplot as plt

# Plot KPI1 over time
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['KPI1'], label='KPI1', marker='o')
plt.title("Trend of KPI1 Over Time")
plt.xlabel("Date")
plt.ylabel("KPI1")
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()

# Histogram of KPI2
plt.figure(figsize=(8, 4))
plt.hist(df['KPI2'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of KPI2")
plt.xlabel("KPI2")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

## Knowledge Check: Perform Your Own Analysis

1. Create a line chart for `KPI3` over time.
2. Calculate and print the correlation matrix for the KPIs.
3. Plot a box plot to observe the distribution of `KPI4` and `KPI5`.

```python
# Problem 1: Line chart for KPI3 over time
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['KPI3'], label='KPI3', marker='s', color='green')
plt.title("Trend of KPI3 Over Time")
plt.xlabel("Date")
plt.ylabel("KPI3")
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()

# Problem 2: Correlation matrix
correlation_matrix = df.iloc[:, 1:].corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Problem 3: Box plot for KPI4 and KPI5
plt.figure(figsize=(8, 5))
df[['KPI4', 'KPI5']].boxplot()
plt.title("Box Plot for KPI4 and KPI5")
plt.ylabel("Values")
plt.grid()
plt.show()
```
